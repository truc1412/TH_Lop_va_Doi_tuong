```python
""" Viết Class NhanVien gồm các thuộc tính :
+ Tên 
+ Tuổi 
+ Địa chỉ 
+ Tổng số giờ làm

Và các phương thức :
- Phương thức tạo
- input_info(): Nhập các thông tin cho nhân viên từ bàn phím
- print_info(): In ra tất cả các thông tin của nhân viên
- tinh_thuong(): Tính toán và trả về số tiền thưởng của nhân viên theo công thức sau :
Nếu tổng số giờ làm của nhân viên >= 200 thì thưởng = lương * 20%
Nếu tổng số giờ làm của nhân viên <200  và >= 100 thì thưởng = lương * 10%
Nếu tổng số giờ làm của nhân viên <100 thì thưởng = 0"""
```

```python
class Nhanvien:
    def _init_(self,ten,tuoi,dia_chi,tsgl):
        self.ten=''
        self.tuoi=''
        self.dia_chi=''
        self.tsgl=''
    
    def input_info(self):
        self.ten=input("Nhập tên: ")
        self.tuoi=input("Nhập tuổi: ")
        self.dia_chi=input("Nhập địa chỉ: ")
        self.tsgl=int(input("Nhập tổng số giờ làm: "))
        
    def print_info(self):
        print("Tên:",self.ten)
        print("Tuổi:",self.tuoi)
        print("Địa chỉ:",self.dia_chi)
        print("Tổng số giờ làm:",self.tsgl)
    
    def tinh_thuong(self):
        luong=float(input("Tiền lương: "))
        if self.tsgl >= 200:
            print("Tiền thường là:",luong*0.2)
        elif self.tsgl < 200 and self.tsgl >=100:
            print("Tiền thường là:",luong*0.1)
        else:
            print("Tiền thường là:",luong*0)
        
        
    truc=Nhanvien()
    truc.input_info()
    truc.print_info()
    truc.tinh_thuong()
```

```python
"""Tạo Class có thên SoHoc gồm có các thuộc tính và phương thức sau :
+ Thuộc tính : number1 ,number2
+ Phương thức :
- phương thức tạo __int()__
- input_info(): dùng để nhập 2 số number1 và number2
- print_info(): dùng để hiển number1 và number2
- addition(): dùng để cộng number1,number2
- subtract(): dùng để trừ number1 và number2
- multi(): dùng để nhân number1 và number2 
- division(): dùng để chia number1 và number2

Sau khi tạo class , khởi tạo 1 object SoHoc và in ra kết quả tính toán từ các hàm thực hiên phép tính"""
```

```python
class SoHoc:
   
    def _int_(self,number1,number2):
        self.number1=number1
        self.number2=number2
    def input_info(self):
        self.number1=int(input("Nhập số thứ nhất: "))
        self.number2=int(input("Nhập số thứ hai: "))
        
    def print_info(self):
        print("Số thứ nhất là:",self.number1)
        print("Số thứ hai là:",self.number2)
        
    def addition(self):
        print("Tổng của 2 số là: ",self.number1+self.number2)
        
    def subtract(self):
        print("Hiệu của 2 số là: ",self.number1-self.number2)
        
    def multi(self):
        print("Tích của 2 số là: ",self.number1*self.number2)
        
    def division(self):
        if self.number2 == 0:
            print("Không chia được")
        else:
            print("Thương của 2 số là: ",self.number1/self.number2)

so_hoc=SoHoc()
so_hoc.input_info()
so_hoc.print_info()
so_hoc.addition()
so_hoc.subtract()
so_hoc.multi()
so_hoc.division()
```

```python
"""Hãy tạo class SinhVien gồm các thông tin: họ tên, mã số, ngày sinh, địa chỉ, điểm
trung bình 4 năm học, số điện thoại. Mỗi sinh viên có thể có tối đa 3 số điện thoại.
Viết chương trình cho phép:
- Nhập một danh sách các sinh viên với đầy đủ các thông tin nêu trên.
- Thêm một sinh viên vào danh sách.
- Tìm kiếm sinh viên trong danh sách theo mã số hoặc theo họ tên.
- Xóa một sinh viên khỏi danh sách dựa theo mã số.
- Sửa dữ liệu một sinh viên.
- Sắp xếp lại các sinh viên theo họ tên hoặc theo mã số.
- Sắp xếp lại các sinh viên theo điểm trung bình từ cao đến thấp.
- Xuất danh theo thứ tự mã số tăng dần.
- Xuất danh sách theo thứ tự điểm trung bình giảm dần và có thêm cột sắp hạng
sinh viên. 
 """
```

```python
class SinhVien:
    def _init_(self,name,mssv,ddy,address,gpa,sdt1,sdt2,sdt3):
        self.name=name
        self.mssv=mssv
        self.ddy=ddy
        self.address=address
        self.gpa=gpa
        self.sdt1=sdt1
        self.sdt2=sdt2
        self.sdt3=sdt3
    
def add_sv(lst):
    name=input("Nhập họ và tên: ")
    mssv= input("Nhập MSSV: ")
    ddy=input("Nhập ngày tháng năm sinh: ")
    address= input("Nhập địa chỉ: ")
    gpa= input("Nhập điểm trung bình : ")
    sdt1= input("Nhập stđ 1: ")
    sdt2= input("Nhập stđ 2 nếu không nhập 'n': ")
    if sdt2 == 'n':
        sdt2 = "Không có"
        sdt3 = "Không có"
    else:
        sdt3 = input("Nhập stđ 3 nếu không nhập 'no': ")
        if sdt3=='n':
            sdt3= "Không có"
    sv = SinhVien(name,mssv,ddy,address,gpa,sdt1,sdt2,sdt3)
    lst.append(sv)

def find_sv(lst,name):
    check=False
    for i in range(len(lst)):
        if lst[i].name==name or lst[i].mssv==name:
            lst[i].show()
            check=True
    if check==False:
        print("Sinh viên không có trong danh sách")
            
def del_sv(lst,mssv):
    check=False
    for i in range(len(lst)):
        if lst[i].mssv==mssv:
            check=True
            print("Bạn đang xóa SV", lst[i].name)
            while True:
                choice=input("Bạn có muồn tiếp tục không? (y/n): ")
                if choice.lower() in ['y','n']:
                    break
                else:
                    print("Bạn nhập sai cú pháp!")
            if choice.lower() == 'y':
                del lst[i]
                print("Đã xóa")
    if check==False:
        print("Sinh viên không có trong danh sách")

def adjust_sv(lst,mssv):
    check=False
    for i in range(len(lst)):
        if lst[i].mssv==mssv:
            print("Bạn đang chỉnh sửa thông tin sinh viên", lst[i].name)
            sv=lst[i]
            check=True
            break
    if check == False:
        print("Sinh viên không có trong danh sách")
    else:
        while True: 
            choice=input("Bạn có muồn tiếp tục không? (y/n): ")
            if choice.lower() in ['y','n']:
                break
            else:
                print("'Bạn nhập sai cú pháp!")
        if choice=='y':
                sv.name=input("Nhập họ và tên: ")
                sv.mssv= input("Nhập MSSV: ")
                sv.ddy=input("Nhập ngày tháng năm sinh: ")
                sv.address= input("Nhập địa chỉ: ")
                sv.gpa= input("Nhập điểm trung bình : ")
                sv.sdt1= input("Nhập stđ 1: ")
                sv.sdt2= input("Nhập stđ 2 nếu không nhập 'no': ")
                if sv.sdt2 == 'no':
                    sv.sdt2= "Không có"
                    sv.sdt3= "Không có"
                else:
                    sv.sdt3= input("Nhập stđ 3 nếu không nhập 'no': ")
                    if sv.sdt3 =='no':
                        sv.sdt3 = "Không có"

def sx_lst_mssv(lst):
    sorted(lst,key=lambda x:x.mssv)

def sx_lst_gpa(lst):
    sorted(lst,key=lambda x:x.gpa, reverse=reverse)

def show_lst(lst):
    for i in range(len(lst)):
        lst[i].show()
        
def con():
    global run
    while True:
        choice=input("Bạn có muốn tiếp tục không? (y/n)")
        if choice.lower() == 'n':
            run=False
            break
        elif choice.lower()=='y':
            break
        else:
            print("Bạn nhập sai cú pháp")
            
            
lst_sv=[]
run=True
while run:
    print('Option: \n\
    1. Thêm SV vào danh sách \n\
    2. Tìm kiếm SV theo MSSV hoặc họ và tên \n\
    3. Xóa SV khỏi danh sách theo MSSV\n\
    4. Sửa dữ liệu SV theo MSSV\n\
    5. Xuất danh sách theo MSSV tăng dần\n\
    6. Xuất danh sách theo GPA tăng dần\n\
    7. Xuất danh sách theo GPA giảm dần\n\
    8. Thoát')
    choice=input("Lựa chọn của bạn: ")
    if choice == '1':
        add_sv(lst_sv)
        con()
    elif choice == '2':
        name=input("Nhập tên hoặc MSSV của SV mà bạn muốn tìm: ")
        find_sv(lst_sv,name)
        con()
    elif choice == '3':
        mssv=input("Nhập MSSV của SV mà bạn muốn xóa: ")
        del_sv(lst_sv,name)
        con()
    elif choice == '4':
        name=input("Nhập MSSV của SV mà bạn muốn chỉnh sửa: ")
        find_sv(lst_sv,name)
        con()
    elif choice == '5':
        sx_lst_mssv(lst_sv)
        show_lst(lst_sv)
        con()
    elif choice == '6':
        sx_lst_mssv(lst_sv)
        show_lst(lst_sv)
        con()
    elif choice == '7':
        sx_lst_gpa(lst_sv)
        show_lst(lst_sv)
        con()
    elif choice == '8':
        run= False
    else:
        print("Bạn nhập sai cú pháp")
```

```python
"""
Xây dựng lớp Phân số (Fraction) có các thành phần sau:
- Các thuộc tính: Tử số, mẫu số.
- Phương thức tạo để khởi tạo giá trị cho tử số và mẫu số.
- Các phương thức setter và getter.
- Các phương thức:
    + Nhập phân số
    + In Phân số
    + Rút gọn phân số
    + Nghịch đảo phân số
    + add(), sub(), mul(), div() tương ứng để thực hiện cộng, trừ, nhân, chia hai phân số cho nhau.
"""
```

```python
class Fraction:
    
    def __init__ (self, tuso, mauso):
        self.tuso = tuso
        self.mauso = mauso
    
    def set_tuso(self,tuso):
        self.tuso = tuso
    
    def set_mauso(self,mauso):
        self.mauso=mauso
    
    def get_tuso(self):
        return self.tuso
    
    def get_mauso(self):
        return self.mauso
    
    def input_so (self):
        self.tuso = int(input("Nhap tu so: "))
        self.mauso = int(input("Nhap mau so: "))
    
    def __str__(self):
        return f'{self.tuso}/{self.mauso}'
    
    def rut_gon (self):
        def gcd(a,b):
            while b:
                a,b=b,a%b
            return a
        n=gcd(self.tuso,self.mauso)
        self.tuso //= n 
        self.mauso //=n
        print(f"{self.tuso}/{self.mauso}")
    
    def nghich_dao(self):
        self.tuso, self.mauso = self.mauso, self.tuso
    
    def add(self,other):
        tuso= self.tuso * other.get_mauso() + self.mauso * other.get_tuso()
        mauso= self.mauso * other.get_mauso()
        ket_qua= Fraction(tuso,mauso)
        ket_qua.rut_gon()
        return ket_qua
    
    def sub(self,other):
        tuso= self.tuso * other.get_mauso() - self.mauso * other.get_tuso()
        mauso= self.mauso * other.get_mauso()
        ket_qua= Fraction(tuso,mauso)
        ket_qua.rut_gon()
        return ket_qua
    
    def mul(self,other):
        tuso=self.tuso * other.get_tuso()
        mauso=self.mauso *other.get_mauso()
        ket_qua= Fraction(tuso,mauso)
        ket_qua.rut_gon()
        return ket_qua
    
    def div(self,other):
        tuso=self.tuso *other.get_mauso()
        mauso=self.mauso *other.get_tuso()
        ket_qua= Fraction(tuso,mauso)
        ket_qua.rut_gon()
        return ket_qua
phanso=Fraction(4,5)
phanso.input_so()
phanso.rut_gon()
```

```python
"""Xây dựng lớp hình chữ nhật (Rectangle) có các thành phần sau:
* Các thuộc tính: chiều dài, chiều rộng.
* Các phương thức:
- Nhập chiều dài, chiều rộng
- Tính diện tích
- Tính chu vi"""
```

```python
class Rectangle:
    def _init_(self,cd,cr):
        self.cd=cd
        self.cr=cr
    def nhap_hcn(self):
        self.cd= int(input("Nhập chiều dài: "))
        self.cr= int(input("Nhập chiều rộng: "))
    
    def dt(self):
        print("Diện tích:",self.cd*self.cr)
    
    def cv(self):
       print("Chu vi:",(self.cd+self.cr)*2)
        
cn=Rectangle()
cn.nhap_hcn()
cn.dt()
cn.cv()
```
